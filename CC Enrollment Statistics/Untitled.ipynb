```python
from bs4 import BeautifulSoup
import requests
import time

ctgoodjobs = 'https://www.ctgoodjobs.hk'

all_jobs = []

for page in range(1,112):
    link = f'https://www.ctgoodjobs.hk/ctjob/listing/joblist.asp?job_area=021_jc&page={page}'
    response = requests.get(link).text
    doc = BeautifulSoup(response, 'html.parser')

    jobs = doc.find_all(class_='row jl-row jl-de')
    for job in jobs:
        item = {
            'job_title': job.find(class_='jl-title').a.string,
            'job_link': ctgoodjobs + str(job.find(class_='jl-title').a.get('href', None)),
            'company' :job.find(class_='jl-comp').a.string,
            'location': job.find(class_='jl-job-info').find('li', {'class':'loc col-xs-12'}).get_text(strip=True),
            'experience': job.find(class_='jl-job-info').find('li', {'class':'exp col-xs-12'}).get_text(strip=True),
            'post_date': job.find(class_='jl-job-info').find('li', {'class':'post-date col-xs-12'}).get_text(strip=True),
        }

    all_jobs.append(item)

pd.DataFrame(all_jobs).to_csv('CTGoodJobs_IT.csv')
```
