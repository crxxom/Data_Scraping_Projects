```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://www.futhead.com/23/nations/'
page = requests.get(url)
nation_doc = BeautifulSoup(page.content, 'html.parser')
pages = int(nation_doc.find_all(class_='font-12 font-bold margin-l-r-10')[0].text.strip().split(' ')[-1])
futhead_team = {}
for page in range(1, pages+1):
    url = f"https://www.futhead.com/23/nations/?page={page}"
    page = requests.get(url)
    nation_doc = BeautifulSoup(page.content, 'html.parser')
    
    for nation in nation_doc.find_all(class_='list-group-item list-group-table-row player-group-item dark-hover'):
        img_url = str(nation.img).split(' ')[-1].split('"')[1]
        team_name = nation.find(class_='player-name').text
        futhead_team[team_name] = img_url
        
        
futhead_teams_df = pd.DataFrame({'teams':futhead_team.keys()}, index=list(futhead_team.values()))

futhead_teams_df.loc['https://futhead.cursecdn.com/static/img/22/nations/43.png'] = 'Slovakia'
futhead_teams_df.loc['https://futhead.cursecdn.com/static/img/22/nations/123.png'] = 'Libya'
futhead_teams_df.loc['https://futhead.cursecdn.com/static/img/22/nations/119.png'] = 'Guinea-Bissau'
futhead_teams_df.loc['https://futhead.cursecdn.com/static/img/22/nations/87.png'] = 'Panama'
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://www.futhead.com/23/players/?page=1'
page = requests.get(url)
doc = BeautifulSoup(page.content, 'html.parser')
pages = int(str(doc.find(class_='font-12 font-bold margin-l-r-10')).split('<')[-2].strip().split(' ')[-1])


name_ = []
rating_ = []
card_type_ = []
position_ = []
league_ = []
team_ = []
nation_ = []
PAC_ = []
SHO_ = []
PAS_ = []
DRI_ = []
DEF_ = []
PHY_ = []

for page in range(1, pages+1):
    print(page, end=' ')
    url = f"https://www.futhead.com/23/players/?page={page}"
    page = requests.get(url)
    doc = BeautifulSoup(page.content, 'html.parser')
    all_players = doc.find_all(class_='list-group-item list-group-table-row player-group-item dark-hover')
    for player in all_players:
        name = str(player.find(class_='player-name')).split('<')[-2].split('>')[-1]
        rating = int(str(player.span).split('/')[0].split('>')[-1][:-1])
        card_type = str(player.span).split('"')[-2].split(' ')[-1]
        position = player.find(class_='player-club-league-name').text.split("|")[0].strip()
        try:
            league = player.find(class_='player-club-league-name').text.split("|")[2].strip()
        except:
            league = None
        try:
            team = player.find(class_='player-club-league-name').text.split("|")[1].strip()
        except: 
            team = None
#         nation = futhead_teams_df.loc[str(player.find(class_='player-nation')).split(' ')[3].split('=')[-1].split('"')[1]].teams
        PAC = int(player.find_all(class_='player-stat stream-col-60 hidden-md hidden-sm')[0].span.text)
        SHO = int(player.find_all(class_='player-stat stream-col-60 hidden-md hidden-sm')[1].span.text)
        PAS = int(player.find_all(class_='player-stat stream-col-60 hidden-md hidden-sm')[2].span.text)
        DRI = int(player.find_all(class_='player-stat stream-col-60 hidden-md hidden-sm')[3].span.text)
        DEF = int(player.find_all(class_='player-stat stream-col-60 hidden-md hidden-sm')[4].span.text)
        PHY = int(player.find_all(class_='player-stat stream-col-60 hidden-md hidden-sm')[5].span.text)

        name_.append(name)
        rating_.append(rating)
        card_type_.append(card_type)
        position_.append(position)
        league_.append(league)
        team_.append(team)
#         nation_.append(nation)
        PAC_.append(PAC)
        SHO_.append(SHO)
        PAS_.append(PAS)
        DRI_.append(DRI)
        DEF_.append(DEF)
        PHY_.append(PHY)
        
        
```
